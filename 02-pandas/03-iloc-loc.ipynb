```python
import numpy as np
import pandas as pd
```

There are some ways to select data in pandas DataFrames, the most powerful way is using **iloc** and **loc**.

+ **iloc**: Allow us extract information from the DataFrame by specifying the index, that is what "i" means.

+ **loc**: Acces to data by specifying the label

```python
path = "./Online_Retail.csv"
retail_csv = pd.read_csv(path, encoding="windows-1252")

retail_csv.head()
```

```python
# get the first row
retail_csv.iloc[0]
```

```python
# First 5 rows with iloc
retail_csv[:5]
```

```python
# rows 7 to 11 with iloc
retail_csv[7:12]
```

```python
retail_csv.iloc[10:21, ::2]
```

```python
retail_csv.iloc[10, 4]
```

### loc

```python
# Selecting rows with loc
retail_csv.loc[3]
```

```python
retail_csv.loc[4:9]
```

```python
retail_csv.loc[5:11, "Quantity"]
```

```python
retail_csv.loc[55:61, ["Quantity", "UnitPrice"]]
```
