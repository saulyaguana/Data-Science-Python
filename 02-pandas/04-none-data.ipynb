It's comon to find dataset with none values, this can be happend because some things:

+ Errors in recolection data
+ Storage problems
+ In the moment that were recolecting the data, the data was not available

Identify and managing this none data it's crucial to get an accurate and significant analysis.

There is multiple ways to identify if there is or there is not data, in pandas there are two main ways.

```python
path = "./Online_Retail.csv"
retail_data = pd.read_csv(path, encoding="windows-1252")
retail_data.head()
```

```python
# missing data
retail_data.isna().head()
```

```python
# Counting the missing data by column
retail_data.isna().sum()
```

Now we know there is missing data, we can do two things

1. Delete all rows that has missing data
2. store some value by making some change

```python
# 1. Delete rows with missing data | It can also be the columns with missing data
retail_data.dropna()
```

```python
# Delete missing data by columns
retail_data.dropna(axis=1)
```

```python
# Storign 0 in the missing values
retail_data.fillna(0).isna().sum()
```

```python
retail_data["UnitPrice"].fillna(retail_data["UnitPrice"].mean()).isna().sum()
```
