```python
import pandas as pd
```

```python
df = pd.read_csv("./Online_Retail.csv", encoding="windows-1252")
df.head()
```

```python
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
df.head()
```

```python
df["HighValue"] = df["TotalPrice"] > 100
df["HighValue"].head()
```

```python
df["HighValue"].iloc[500:505]
```

```python
df.info()
```

```python
# data types in a column
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df.info()
```

```python
# lambda
df["DiscountPrice"] = df["UnitPrice"].apply(lambda x: x * 0.9)
df.head(3)
```

```python
def categorize_price(price):
    if price > 50:
        return "High"
    elif price < 20:
        return "Medium"
    return "Low"
```

```python
df["PriceCategory"] = df["UnitPrice"].apply(categorize_price)
df.head(3)
```
