```python
import pandas as pd
```

```python
df = pd.read_csv("./Online_Retail.csv", encoding="windows-1252")
df.head(3)
```

```python
# Show how many times each unique value is in there
df["Country"].value_counts()
```

```python
df.groupby("Country")["Quantity"].sum()
```

```python
df.groupby("Country")["UnitPrice"].agg(["mean", "sum"])
```

```python
df.groupby(["Country", "StockCode"])["Quantity"].sum()
```

```python
def total_revenue(group):
    return (group["Quantity"] * group["UnitPrice"]).sum()
```

```python
df.groupby("Country").apply(total_revenue)
```
