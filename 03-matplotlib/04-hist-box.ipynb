```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
data = np.random.normal(170, 10, 200)
data
```

```python
plt.hist(data, color="deeppink", edgecolor="blue", alpha=1, bins=20)

plt.title("High distribution")
plt.xlabel("Heigh (cm)")
plt.ylabel("Density")

plt.show()
```

# Boxplot

```python
np.random.seed(0)
ages = [
    np.random.normal(30, 5, 100),
    np.random.normal(40, 5, 100),
    np.random.normal(35, 5, 100)
]
ages
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, labels=["Group 1", "Grpup 2", "Group 3"])

plt.title("Age distribution by group")
plt.xlabel("Group")
plt.ylabel("Age (years)")

plt.show()
```
