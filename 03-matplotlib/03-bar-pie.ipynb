```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# config the size
plt.figure(figsize=(8, 6))
```

```python
# data for bar
categories = ["Electronics", "Sports", "Software"]
sales = [1500, 897, 2889]
```

```python
# bar
plt.bar(categories, sales, color="skyblue", label="Month sales")
plt.annotate("Pick", xy=("Software", 897), xytext=("Electronics", 1500), arrowprops=dict(facecolor="black", shrink=0.05))

plt.title("Product sales in a month")
plt.xlabel("Categories")
plt.ylabel("Sales")
plt.legend()

plt.show()
```

```python
plt.barh(categories, sales, color="hotpink", label="Month sales")

plt.title("Products sales in a month")
plt.ylabel("Categories")
plt.xlabel("Amount")

plt.show()
```

# Pie

```python
products = ["Electronics", "Sports", "Software"]
sales = [50, 35, 15]

colors = ["gold", "lightcoral", "orange"]
```

```python
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=140, colors=colors)
plt.axis('equal')

plt.show()
```
