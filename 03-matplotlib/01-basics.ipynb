```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
months = np.array(
    [
        "January",
        "February",
        "March",
        "April",
        "May",
        "June",
        "July",
        "August",
        "September",
        "October",
        "November",
        "Devember",
    ]
)

sales = np.array([9, 12, 10, 15, 16, 23, 21, 27, 38, 28, 39, 42])
```

```python
# config
plt.figure(figsize=(8, 6))
```

```python
# create the graphic
plt.plot(months, sales, marker='o', color='blue')

plt.title('Amount for sales per month')
plt.xlabel("Months")
plt.ylabel("Amount")

plt.show()
```

```python
# data for scatterplot
hours = [2, 3, 4, 5, 6, 7, 8]
score = [55, 60, 65, 70, 75, 80, 85]
```

```python
plt.scatter(hours, score, color='green')

plt.title("Relationship between studying hours and score")
plt.xlabel("Hours")
plt.ylabel("Score")

plt.show()
```
