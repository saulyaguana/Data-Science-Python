```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
dates = pd.date_range(start="2023-01-01", periods=100)
values = np.random.rand(100).cumsum()

data = pd.DataFrame(
    {
        "Date": dates,
        "Value": values,
    }
)
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data["Date"], data["Value"], color="#1c527a")

plt.xticks(rotation=45)
plt.title("Time series with date format")
plt.xlabel("Date")
plt.ylabel("Value")

plt.show()
```

```python
dates = pd.date_range(start="2023-01-01", periods=12, freq='M')
sales = np.random.randint(1000, 5000, size=12)

sales_data = pd.DataFrame(
    {
        "Data": dates,
        "Sale": sales,
    }
)
```

```python
sales_data.head()
```

```python
plt.plot(sales_data["Data"], sales_data["Sale"], marker="o", color="#37bd26", linestyle="-", label="Monthly sales")
plt.gca().axis.set._major_formatter(DateFormatter("%b %Y"))
plt.xticks(rotation=45)

plt.title("Monthly sales analysis")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.legend()
plt.tight_layout()

plt.show()
```
