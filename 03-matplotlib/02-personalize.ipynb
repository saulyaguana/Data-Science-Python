```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# config the size
plt.figure(figsize=(8, 6))
```

```python
# data
hours = [2 ,3, 4, 5, 6, 7, 8]
score_1 = [55, 60, 65, 70, 75, 80, 85]
score_2 = [50, 58, 63, 69, 74, 78, 83]
```

```python
plt.scatter(hours, score_1, marker='o', color='red', linestyle='-', label='Score 1')
plt.scatter(hours, score_2, marker='s', color='blue', linestyle='dashed', label="Score 2")

plt.title("Relationship between studying hours and score")
plt.xlabel("Hours")
plt.ylabel("Score")

plt.show()
```

```python
# plot
plt.plot(hours, score_1, marker='o', color='#f700e7', linewidth=2, linestyle='-', label='Score 1')
plt.plot(hours, score_2, marker='s', color='#00daf7', linewidth=2, linestyle='dashed', label='Score 2')

plt.show()
```
